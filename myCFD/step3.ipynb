```python
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
nx = 41
dx = 2 / (nx - 1)
nt = 100
nu = 0.3
sigma = 0.2
dt = sigma * dx**2 / nu

u = np.ones(nx)
u[int(0.5 / dx):int(1 / dx + 1)] = 2

un = np.ones(nx)
```

```python
for n in range(nt):
    un = u.copy()
    u[1:-1] = un[1:-1] + nu * dt / dx**2 * (un[2:] - 2 * un[1:-1] + un[:-2])    
    plt.plot(np.linspace(0, 2, nx), u);
```
